**Sine-Function QML**

In this code, we build a Quantum Neural Network to approximate the sine function. We first discretize the interval [0,2ùúã] into 100 points and compute the sine values as labels. The QuantumLayer integrates a quantum circuit with angle embedding and entangling layers. The model is trained using mean squared error to minimize the difference between predicted and actual sine values. The quantum layer processes the input features and learns to approximate the sine function using quantum operations.




```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt

# Define the quantum device
dev = qml.device("default.qubit", wires=3)

# Define the QML model
@qml.qnode(dev)
def qml_model(phi, weights):
    qml.RX(phi, wires=0)
    qml.RY(weights[0], wires=0)
    qml.CNOT(wires=[0, 1])
    qml.RY(weights[1], wires=1)
    qml.CNOT(wires=[1, 2])
    qml.RY(weights[2], wires=2)
    return qml.expval(qml.PauliZ(0))

# Define the sine function
def sine(x):
    return np.sin(x)

# Define the training data
x_train = np.linspace(0, 2 * np.pi, 200)
y_train = np.sin(x_train)

# Define the loss function
def loss(weights):
    predictions = np.array([qml_model(x, weights) for x in x_train])
    return np.mean((predictions - y_train) ** 2)

# Train the QML model
opt = qml.AdamOptimizer(0.01)
weights = np.random.normal(0, 0.1, (3,), requires_grad=True)

for epoch in range(200):
    weights, _ = opt.step_and_cost(lambda w: loss(w), weights)

# Evaluate the trained QML model
x_eval = np.linspace(0, 2 * np.pi, 200)
y_eval = np.array([qml_model(x, weights) for x in x_eval])

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x_train, y_train, label="Sine function")
plt.plot(x_eval, y_eval, label="QML model", linestyle='--')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Sine Function Approximation using Quantum Machine Learning")
plt.grid(True)
plt.show()
```
